### 1. 행렬과 벡터의 곱셈

##### 가상의 환자 데이터 행렬
##### 행: 환자 3명, 열: [소득분위, 나이, BMI]
##### 종속 변수로 회귀계수를 넣어 수축기 혈압 예측치 구하기

```python
import numpy as np

X = np.array([
    [1, 45, 22],
    [2, 32, 28],
    [3, 55, 30]
])

print(X.shape)

y = np.array([0.8, 1.2, 1.8])

print(y.shape)    # 행이 3개인 1차원 벡터
```

##### 행렬 X와 벡터 y의 행렬 곱(앞 행렬의 열 개수와 뒤 행렬의 행 개수가 반드시 같아야)

```python
z=X@y

print("환자별 예측 수축기 혈압:", z)
```

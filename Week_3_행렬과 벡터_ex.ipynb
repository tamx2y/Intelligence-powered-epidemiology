### 0. 행 벡터와 열 벡터 Vectors from a Matrix

```python
import numpy as np

A = np.array([[1, 6], [3, 4], [5, 2]])
print("--- 원본 행렬 A ---")
print(A)

# 1. 행 벡터 (Row Vector) 추출
# 파이썬은 0부터 숫자를 세므로, 2번째 행은 인덱스 [1]

row_vector = A[1, :]
print("\n--- 2번째 행 벡터  ---")
print(row_vector)

# 2. 열 벡터 (Column Vector) 추출
# 2번째 열 역시 인덱스 [1]을 사용

column_vector = A[:, 1]
print("\n--- 2번째 열 벡터 ---")
print(column_vector)
```

### 1. 행렬과 벡터의 곱셈

##### 가상의 환자 데이터 행렬
##### 행: 환자 3명, 열: [소득분위, 나이, BMI]
##### 종속 변수로 회귀계수를 넣어 수축기 혈압 예측치 구하기

```python
X = np.array([
    [1, 45, 22],
    [2, 32, 28],
    [3, 55, 30]
])

print(X.shape)

y = np.array([0.8, 1.2, 1.8])

print(y.shape)    # 행이 3개인 1차원 벡터
```

##### 행렬 X와 벡터 y의 행렬 곱(앞 행렬의 열 개수와 뒤 행렬의 행 개수가 반드시 같아야)

```python
z=X@y

print("환자별 예측 수축기 혈압:", z)
```
